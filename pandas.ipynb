Video 3: Filtering

```python
filt = (people['last'] == "Doe") # get a series object
```

```python
people[filt] # get a dataframe back with rows with last name doe
```

```python
people[people['last'] == "Doe"] # can also put the filter directly in
```

```python
people.loc[filt, 'email'] # second value columns you want
```

```python
# & and | or
filt = (people['last'] == "Doe") & (people['first'] == 'John')
```

```python
people.loc[filt]
```

```python
filt2 = (people['last'] == "Schafer") | (people['first'] == 'John')
# last name Schafer or first name John
people.loc[filt2]
```

```python
# last name isn't Schafer and first name isn't John
people.loc[~filt2] # just put ~ in front of filt
```

```python
# can use string methods in Python
filt = df['LanguageWorkedWith'].str.contains("Python", na = False)
df.loc[filt, 'LanguageWorkedWith']
# results contain Python
```

Part 5: Updating Data

```python
people.columns
```

```python
# remain all columns using the .columns 
people.columns = ['email', 'first_name', 'last_name']
```

```python
# more common to change something specific about each column in df
# all uppercase
people.columns = [x.upper() for x in people.columns]
people
```

```python
# replace spaces with underscores and vice versa
people.columns = people.columns.str.replace(' ' , '_')
people
```

```python
people.columns = [x.lower() for x in people.columns]
```

```python
people.rename(columns = {'first_name': 'first', 'last_name': 'last', 'email': 'email'}, inplace = True)
```

```python
people
```

```python
# updating data in rows
# imagine a lot of columns and only want to change a couple of values
people.loc[2, ['last', 'email']] = ['Smith', 'johnsmith@email.com']
people
```

```python
people.loc[2, 'last'] = 'Doe'
people
```

```python
# can also use at instead of loc or iloc
people.at[2, 'last'] = 'Smith'
people
```

```python
people
```

```python
people['email'] = people['email'].str.lower()
people
```

```python
people['email'].apply(len)
```

```python
def update_email(email):
    return email.upper()
```

```python
people['email'] = people['email'].apply(update_email)
```

```python
people
```

```python
people['email'] = people['email'].apply(lambda x: x.lower())
```
